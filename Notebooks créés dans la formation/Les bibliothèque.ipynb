Petit script

```python
import requests as rq

# définir l'url
url = 'https://www.lemonde.fr/'

# récupérer la page
page = rq.get(url)

contenu_page = page.text

# écrire dans un fichier
with open("page_monde.html","w",encoding="utf8") as file:
    file.write(contenu_page)
```

Montée en abstraction

```python
# Données
liste_urls = ['https://www.lemonde.fr/',
              "https://www.liberation.fr/",
              "https://www.lefigaro.fr/"]

# Définir un compteur
compteur = 0

# Boucle sur la liste
for url in liste_urls:
    print(url)
    page = rq.get(url)
    contenu_page = page.text
    with open(f"fichier{compteur}.html","w",encoding="utf8") as file:
        file.write(contenu_page)
    compteur+=1
```

```python
# Données
dico_urls = {"Le Monde":'https://www.lemonde.fr/',
              "Libération":"https://www.liberation.fr/",
              "Le Figaro":"https://www.lefigaro.fr/"}

# Boucle sur la liste
for n in dico_urls:
    print(n)
    page = rq.get(dico_urls[n])
    contenu_page = page.text
    
    with open(f"{n}.html","w",encoding="utf8") as file:
        file.write(contenu_page)
        
    print(contenu_page.lower().count("covid"))
print("C'est fini")
```

## Construire une fonction

une fonction qui prend en entrée un texte et un mot-clé et qui compte le nombre d'occurence du mot clé en minuscule

```python
def compter_motcle(texte,motcle):
    texte_minuscule = texte.lower()
    nbre_motcle = texte_minuscule.count(motcle)
    return nbre_motcle

def compter_motcle(texte,motcle):
    return texte.lower().count(motcle)
```
